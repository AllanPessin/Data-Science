```
# Metodo Scartterplot
x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]

titulo = "Scatterplot: gráfico de disperção"
eixox = "Eixo X"
eixoy = "Eico Y"

#Titulo
plt.title(titulo)#title define o nome do eixo
#Eixos
plt.xlabel(eixox)#label define o nome do eixo
plt.ylabel(eixoy)#label define o nome do eixo
 
#.scatter mostra o grafico de disperção
#personalização no .scatter altera os pontos
#label dentro da função .bar indica uma legenda das informações do grafico
#color define a cor dos pontos 
#marker muda o ponto no grafico de acordo com a especificação
#marker = . , v ^ 1 2 34 5 s p * h H  + x D d | 
#atributo 's' define o tamanho do ponto do grafico
plt.scatter(x, y, label = "Meus pontos", color = "green", marker=".", s = 200)

#personalização no .plot altera as linhas
#color define a a cor das linhas
#linestyle deine o tipo da linha
plt.plot(x, y, color = 'black', linestyle = ':')# liga os ponto com uma linha 
plt.legend()# necessário para mostrar a legenda dentro do grafico
plt.show()# mostra apenas os labels dos eixos e titulos
```

```
import matplotlib.pyplot as plt

#Crescimento da população Brasileira de 1980 a 2016
#Com base no DataSus

#Ler arquivo
dados = open("populacao_brasileira.csv").readlines()#open() abre o arquivo de upload

x= []
y= []

for i in range(len(dados)):
  if i != 0:
    linha = dados[i].split(";")
    x.append(int(linha[0]))
    y.append(int(linha[1]))

plt.plot(x, y, color="black", linestyle="--")
plt.bar(x, y, color="cyan")
plt.title("Crescimento da população brasileira de 1980 a 2016")
plt.xlabel("Ano")
plt.ylabel("População em milhões")
#plt.show()
plt.savefig("População brasileira.png")
```

```
# Boxplo
import matplotlib.pyplot as plt
import random #geração de numeros aleatorio

vetor = []

for i in range(10):
  numero_aleatorio = random.randint(0,5)#gera numero aleatorios inteiros
  vetor.append(numero_aleatorio)

plt.boxplot(vetor)
plt.title("Boxplot")
plt.show()
```

```
entrada = open("human.fasta").read()
saida = open("humano.html", "w")

cont={}

for i in ['A', 'T', 'C', 'G']:
  for j in ['A', 'T', 'C', 'G']:
    cont[i+j] = 0
entrada = entrada.replace("\n","")    
for k in range(len(entrada)-1):
  cont[entrada[k]+entrada[k+1]] += 1
print(cont)

#HTML
i = 1
for k in cont:
  transparencia = cont[k]/max(cont.values())
  saida.write("<div style='width:100px; border:1px solid #111; color=#fff; height:100px; float:left; background-color:rgba(0,0,255,"+str(transparencia)+"')>"+k+"</div>")
  if i%4 == 0:
    saida.write("<div style='clear:both'></div>")
  i+=1

saida.close()
```
